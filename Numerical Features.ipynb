#Handling Categorical & Numerical Features

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
```

```python
df = pd.DataFrame({
    "Name"          : ['Sade', 'Bob', 'Charlie', 'Joan', 'Bose', 'Odogwu'],
    "gender"        : ['Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
    "city"          : ['Lagos', 'Abuja', 'Ibadan', 'Lagos', 'Abuja', 'Ibadan'],
    "education"     : ['Graduate', 'Not Graduate', 'Graduate', 'Graduate', 'Not Graduate', 'Graduate'],
    "risk_level"    : ['Low', 'Medium', 'High', 'Low', 'High', 'Medium'],  
    "income"        : [55000, 32000, 78000, 91000, 46000, 63000],
    "loan_amount"   : [200000, 150000, 450000, 500000, 175000, 310000],
    "loan_approved" : [1, 0, 1, 1, 0, 1]
})
df
```

```python
dummies = pd.get_dummies(df['Name'], dtype=int)
dummies
```

```python
merged = pd.concat([df, dummies],axis='columns')
merged
```

```python
new_df = merged.drop(['Name'], axis='columns')
new_df
```

```python
new_df["gender"]
```

```python
label_encoder = LabelEncoder()

new_df["gender"] = label_encoder.fit_transform(new_df["gender"])

new_df
```

```python
new_df[["risk_level"]]
new_df
```

```python
ord_encoder = OrdinalEncoder()

new_df["risk_level"] = ord_encoder.fit_transform(new_df[["risk_level"]])
new_df
```

```python
df.head()
```

```python
ct = ColumnTransformer([('Name', OneHotEncoder(), [0])], remainder = 'passthrough')
ct_x = ct.fit_transform(df)
ct_x
```

STANDARDIZATION AND NORMALIZATION

```python
# Normalization using MinMaxScaler
scaler = MinMaxScaler()
new_df[['loan_amount', 'loan_approved']] = scaler.fit_transform(new_df[['loan_amount', 'loan_approved']])
new_df
```

```python
# Standardization using StandardScaler
std_scaler = StandardScaler()
new_df[['loan_amount', 'loan_approved']] = std_scaler.fit_transform(new_df[['loan_amount', 'loan_approved']])
new_df
```
