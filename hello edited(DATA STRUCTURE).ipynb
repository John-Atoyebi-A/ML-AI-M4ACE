SIMPLE CALCULATOR

```python
def school_age_calculator(age,name):
    if age< 6:
        print('chop rice', name,'is only', age)
    elif age == 6: 
        print('welcome to school', name)
    else:
        print('they grow up so fast')
school_age_calculator(10,'Joan')
```
